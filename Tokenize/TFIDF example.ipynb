## TFIDF can capture word importance by the present frequency

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download('stopwords')
```

## Load Dataset

```python
message = pd.read_csv('spam.csv', skiprows = [0], usecols=[0, 1], names = ["label","message"], encoding = 'ISO-8859-1')
message
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len(message)):
    review = re.sub("[^a-zA-Z]", " ", message["message"][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(w) for w in review if w not in stopwords.words("english")]
    review = " ".join(review)
    corpus.append(review)

corpus[:20]
```

## Implement TF-IDF and N-grams

```python
tfidf = TfidfVectorizer(max_features = 1000, ngram_range = (1,2))
```

```python
x = tfidf.fit_transform(corpus).toarray()
```

```python
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
```
